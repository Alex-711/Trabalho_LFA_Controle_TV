```python
from automathon import DFA
from visual_automata.fa.dfa import VisualDFA
#q0: TV desligada
#q1: TV ligada
#q2: Escolher usuário no aplicativo Netflix
#q3: Escolher filme no aplicativo Netflix
#q4: Escolher filme no aplicativo Globo Play( não requer autenticação)
#q5: Escolher usuário no aplicativo Prime Video
#q6: Escolher filme no aplicativo Prime Video

Q = {'q0', 'q1', 'q2','q3','q4', 'q5','q6'}

#L: Power Button
#K: Enter/OK
#+: Aumentar o volume
#- : Diminuir o volume
#N: Netflix
#P: Prime Video
#G: Globo Play
sigma = {'L', 'K','+','-','N','P','G'}
delta = {
  'q0': {
    '+': 'q0',
    '-': 'q0',
    'N': 'q0',
    'G': 'q0',
    'P': 'q0',
    'K': 'q0',
    'L': 'q1',
  },
  'q1': {
    'L': 'q0',
    '+': 'q1',
    '-': 'q1',
    'K': 'q1',
    'N': 'q2',
    'P': 'q5',
    'G': 'q4',
  },
  'q2': {
    'L': 'q0',
    '+': 'q2',
    '-': 'q2',
    'G': 'q4',
    'N': 'q2',
    'P': 'q5',
    'K': 'q3',
  },
  'q3': {
    'L': 'q0',
    '+': 'q3',
    '-': 'q3',
    'N': 'q2',
    'G': 'q4',
    'P': 'q5',
    'K': 'q3',
  },
  'q4': {
    'L': 'q0',
    'G': 'q4',
    '+': 'q4',
    '-': 'q4',
    'P': 'q5',
    'N': 'q2',
    'K': 'q5',
  },

  'q5': {
    'L': 'q0',
    '+': 'q5',
    '-': 'q5',
    'P': 'q5',
    'N': 'q2',
    'G': 'q4',
    'K': 'q6',
  },
  'q6': {
    'L': 'q0',
    '+': 'q6',
    '-': 'q6',
    'K': 'q6',
    'P': 'q5',
    'N': 'q2',
    'G': 'q4',
  }
}

initial_state = 'q0'
F = {'q3','q4','q6'}

automata = DFA(Q, sigma, delta, initial_state, F)
```

```python
automata.initialState
```

```python
dfa = VisualDFA(states = automata.Q, input_symbols = automata.sigma,transitions=automata.delta,initial_state = automata.initialState,final_states=automata.F)
dfa.table
```
